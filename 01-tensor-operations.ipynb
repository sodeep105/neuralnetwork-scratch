## Function 1 - torch.zeros

This function returs a tensor filled with scalar value of 0 and the shape and size of the tensor has to be passed as arguments. The general structure is: 
torch.zeros(*size, out=None, dtype=None, layout=torch.strided, device=None, requires_grad=False)
Size must be provided for this function to work.

```python
# Example 1 - working 
torch.zeros([5,10])
```
